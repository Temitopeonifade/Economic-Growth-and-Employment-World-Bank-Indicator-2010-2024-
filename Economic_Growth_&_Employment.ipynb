```python
import pandas as pd
df = pd.read_csv('WDICSV.csv')
df.head()
```

```python
for indicator in sorted (df["Indicator Name"].unique()):
  print(indicator)
```

```python
#indicators i want to use for my analysis
indicators = [
    "GDP growth (annual %)",
    "GDP per capita (current US$)",
    "Labor force participation rate, total (% of total population ages 15+) (modeled ILO estimate)",
    "Unemployment, total (% of total labor force) (modeled ILO estimate)",
    "Employment to population ratio, 15+, total (%) (modeled ILO estimate)",
    "Inflation, consumer prices (annual %)",
    "Population, total"
]
```

```python

df[df["Indicator Name"].isin(indicators)]
```

```python
#years of analyzes (2010-2024)
years = [str(year) for year in range (2010, 2025)]
```

```python
#columns and rows needed for analysis
df = df[df["Indicator Name"].isin(indicators)]

columns_to_keep = [
    "Country Name",
    "Country Code",
    "Indicator Name",
    "Indicator Code"
] + years

df = df[columns_to_keep]
```

```python
#changing year columns per year to row
df_long = df.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    var_name='Year',
    value_name='Value'
)
print("Long Format:")
print(df_long.head())
```

```python
df.fillna(0, inplace=True)
print(df.isnull().sum())
```
